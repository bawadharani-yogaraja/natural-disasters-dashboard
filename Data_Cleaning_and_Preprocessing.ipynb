```python
import pandas as pd

# Read the Excel file
df = pd.read_excel('emdat-country-profiles_2026_04_24.xlsx',
                   sheet_name='EM-DAT (2026-04-24)')

# Skip the first row (it's HXL tags, not real data)
df = df.iloc[1:].reset_index(drop=True)

print("✅ File loaded!")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
df.head(10)
```

```python
# Convert number columns from text to actual numbers
numeric_cols = [
    'Year',
    'Total Events',
    'Total Affected',
    'Total Deaths',
    'Total Damage (USD, original)',
    'Total Damage (USD, adjusted)'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("✅ Data types fixed!")
print(df.dtypes)
```

```python
# Drop rows where Year is missing
df = df[df['Year'].notna()]
df['Year'] = df['Year'].astype(int)

print("✅ Bad rows removed!")
print("Remaining rows:", len(df))
```

```python
# Fill blank numbers with 0 (so charts don't break)
df['Total Deaths'] = df['Total Deaths'].fillna(0)
df['Total Affected'] = df['Total Affected'].fillna(0)
df['Total Events'] = df['Total Events'].fillna(0)
df['Total Damage (USD, adjusted)'] = df['Total Damage (USD, adjusted)'].fillna(0)
df['Total Damage (USD, original)'] = df['Total Damage (USD, original)'].fillna(0)

print("✅ Missing values filled!")
print("Any nulls left?")
print(df[numeric_cols].isnull().sum())
```

```python
dupes = df.duplicated().sum()
print(f"Duplicate rows found: {dupes}")

# Remove duplicates if any
df = df.drop_duplicates()
print(f"✅ Clean! Final shape: {df.shape}")
```

```python
print("📊 DATASET SUMMARY")
print("="*40)
print(f"Total rows: {len(df)}")
print(f"Year range: {df['Year'].min()} - {df['Year'].max()}")
print(f"Countries: {df['Country'].nunique()}")
print(f"Disaster Types: {df['Disaster Type'].nunique()}")
print(f"\nDisaster Types:")
print(df['Disaster Type'].value_counts())
```

```python
# Save as CSV
df.to_csv('emdat_cleaned.csv', index=False)

# Download
from google.colab import files
files.download('emdat_cleaned.csv')

print("✅ Clean file downloaded to your computer!")
```
